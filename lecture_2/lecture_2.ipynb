```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
FONTSIZE  = 18
```

$\frac{dS_1}{dt} = -(k_f +k_r)S_1 + k_r S_T$

```python
k_f = 1
k_r = 0.8
S_T = 10
values = np.array([0.1*v for v in range(0, 10*S_T+1)])
def ds(s):
     return np.array(-(k_f + k_r)*s + k_r*S_T)
fig, ax = plt.subplots(1, figsize=(8, 6))
ax.plot(values, ds(values))
ax.set_xlabel(r'$S_1$', fontsize=FONTSIZE)
ax.set_ylabel(r'$\frac{dS_1}{dt}$', fontsize=FONTSIZE, rotation=0)
_ = ax.plot([0, 10], [0, 0], linestyle="--")
```

$\frac{dS_1}{dt} = -(k_f +k_r)S_1(S_T - S_1) + k_r S_T $

```python
k_f = 1
k_r = 0.8
S_T = 10
values = np.array([0.1*v for v in range(0, 10*S_T+1)])
def ds(s):
     return np.array(-(k_f + k_r)*s*(S_T-s) + k_r*S_T)
fig, ax = plt.subplots(1, figsize=(8, 6))
ax.plot(values, ds(values))
ax.set_xlabel(r'$S_1$', fontsize=FONTSIZE)
ax.set_ylabel(r'$\frac{dS_1}{dt}$', fontsize=FONTSIZE, rotation=0)
_ = ax.plot([0, 10], [0, 0], linestyle="--")
```
