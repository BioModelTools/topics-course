# Accessing Files in Google Drive from Colab
This notebook shows how to access files in Google Drive from Colab. You will need to do this for some homeworks that have external data files
that are in the ``Modeling_Class`` shared drive.

## Create a Shortcut
The ``Modeling_Class`` shared folder needs to have a shortcut in your google drive.
1. Browse to ``drive.google.com`` and select (on the left pane) ``Shared with me``. You should see ``Modeling_Class``.
1. Right click on ``Modeling_Class``.
1. Select ``Add shortcut to drive``

Next, open ``MyDrive``, and you should see ``Modeling_Class`` in the list of folders.

## Mount your drive
Run the following cell.
It will generate a link.
Follow the link, copy the code, and paste it into the input field created below.
You will only need to do this one time.

```
from google.colab import drive
drive.mount('/content/drive')
```

## Create a path to the file
Create a variable that is the path to the file you will access.
This is illustrated below for the file ``nst-est2019-01.csv``.

```
filePath = "/content/drive/MyDrive/Winter 2021/Labs/nst-est2019-01.csv"
```

## Access the File
You can now access the filePath as you would a file on your computer. Below is an example of reading a CSV file into a DataFrame.

```
import pandas as pd
df = pd.read_csv(filePath)
df.head()
```
